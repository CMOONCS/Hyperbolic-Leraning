## Standard K-Means Algorithm

Given $x_1, \ldots , x_N \in \mathbb{R}^d$, k-means partitions the $N$ observations into k sets denoted by $\mathbf{S}=\left\{S_{1}, S_{2}, \ldots, S_{k}\right\}$, where we have $k$ centroids or means $\left\{\boldsymbol{m}_{i}^{(t)}\right\}_{i=1}^{k}$ at each iteration $t$. <br>
The basic steps in k-means are: <br>
1. **Initialization step**: initialize K random centroids<br>
2. **Assignment step**: assign each observation $x_p$ to nearest centroid or mean $m_i^{(t)}$ <br>
> $$
S_{i}^{(t)}=\left\{x_{p} :\left\|x_{p}-m_{i}^{(t)}\right\|^{2} \leq\left\|x_{p}-m_{j}^{(t)}\right\|^{2} \forall j, 1 \leq j \leq k\right\}
$$
3. **Update step**: calculate new means (centroids) of the observations in new clusters <br>
> $$
m_{i}^{(t+1)}=\frac{1}{\left|S_{i}^{(t)}\right|} \sum_{x_{j} \in S_{i}^{(t)}} x_{j}
$$ <br>
4. **Repeat until stopping criteria** <br>
5. **Return:** cluster labels for all $x_i$ and cluster means $\left\{\boldsymbol{m}_{i}\right\}_{i=1}^{k}$


To apply k-means in hyperbolic space, we need to have well defined distance functions (which are defined within several models of hyperbolic space). We also need some way to calculate the mean within clusters which doesn't seem to be as straight forward in the poincaré disk, and might require utilizing multiple models of hyperbolic space.

First I start outlining some basic implementations of these core steps.

```python
def generate_data(n, radius=1):
    theta = np.random.uniform(0, 2*np.pi, n)
    u = np.random.uniform(0, radius, n)
    r = np.sqrt(u)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    init_data = np.hstack((x.reshape(-1,1), y.reshape(-1,1)))
    return init_data
```

```python
class PoincareKMeans():
    
    def __init__(self,n_clusters=8,n_init=20,max_iter=300,tol=1e-8,verbose=False):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.max_iter = max_iter
        self.tol = tol
        self.verbose =  verbose
        self.labels_ = None
        self.cluster_centers_ = None
        
    def init_assign(self):
        assignments = np.zeros((self.n_samples, self.n_clusters))
        for i in range(self.n_samples):
            # initialize binary assignment vector
            j = np.random.randint(0, self.n_clusters)
            assignments[i][j] = 1
        self.assignments = assignments
        
    def update_centroids(self, X):
        dim = X.shape[1]
        new_centroids = np.empty((self.n_clusters, dim)) 
        for i in range(self.n_clusters):
            # find total observations in cluster
            n_k = np.sum(self.assignments, axis=0)[i]
            if n_k == 0:
                new_centroids[i] = np.zeros((1, dim))
            else:
                new_centroids[i] = np.sum(X[self.assignments[:, i] == 1], axis=0) / n_k
        self.centroids = new_centroids
    
    def fit(self, X, init=None):
        """
        Uses gradient descent to find the embedding configuration in the Poincaré disk
        Parameters
        ----------
        X : array, shape (n_samples, n_features) or (n_samples, n_samples)
            Input data. If ``dissimilarity=='precomputed'``, the input should
            be the dissimilarity matrix.
        init: initial configuration of the embedding coordinates
        """
        self.fit_transform(X, init=init)
        return self

    def fit_transform(self, X, max_epochs=40, verbose=False):
        """
        Fit the embedding from X, and return the embedding coordinates
        Parameters
        ----------
        X : array, shape (n_samples, n_features) or (n_samples, n_samples)
            Input data. If ``dissimilarity=='precomputed'``, the input should
            be the dissimilarity matrix.
        max_epochs: maximum number of gradient descent iterations
        verbose: optionally print training scores
        """
        
        self.n_samples = X.shape[0]
        
        # initialize random centroid assignments
        self.init_assign()
        
        # loop through the assignment and update steps
        for j in range(max_epochs):
            self.update_centroids(X)
            for i in range(self.n_samples):
                # zero out current cluster assignment
                self.assignments[i, :] = np.zeros((1, self.n_clusters))
                # find closest centroid mean
                centroid_distances = list(np.sqrt(((X[i] - self.centroids)**2).sum(axis=1)))
                cx = centroid_distances.index(np.min(centroid_distances))
                self.assignments[i][cx] = 1
            if verbose:
                print('Epoch ' + str(j) + ' complete')
                print(self.centroids)
        return self.assignments
```

## Adapting to Hyperbolic Space

To implement Poincaré K-Means, we need to develop a method to perform the centroid updating step in hyperbolic space, since the simple arithmetic mean in Euclidean space doesn't translate to hyperbolic space. In **[2]**, the authors propose computing the Fréchet mean, which generalizes the idea of centroids to any complete metric space.

For metric space $(M, d)$, let $x_1, \ldots, x_N$ be points in M. For some point $p \in M$, the Fréchet variance is the sum of squared distances between $p$ and each $x_i$:<br>

$$
\Psi(p)=\sum_{i=1}^{N} d^{2}\left(p, x_{i}\right)
$$ <br>
For each centroid update, we have the optimization problem: <br>

$$
m=\underset{p \in M}{\arg \min } \sum_{i=1}^{N} d^{2}\left(p, x_{i}\right)
$$<br>
It's been shown recently that gradient descent in the Hyperboloid or Lorentz model has significant advantages over optimization in the Poincaré disk **[3]**, so I first attempt utilizing the Hyperboloid model. The following mathematical background and discussion of optimization in the Hyperboloid model is presented very informatively in **[3]**.

## The Hyperboloid Model

The advantages of optimization in the hyperboloid model are similar to the advantages of optimization on an n-dimensional unit sphere - both are able to utilize a higher dimensional ambient space.<br>
<br>
The ambient space for the hyperboloid model is *Minkowski* space, a copy of $\mathbb{R}^{n+1}$ denoted by $\mathbb{R}^{n : 1}$ with the following bilinear form:
$$
\langle u, v\rangle_{n : 1}=\sum_{i=1}^{n} u_{i} v_{i}-u_{n+1} v_{n+1}
$$<br>
The n-dimensional *Hyperboloid* model is defined: <br>
<br>
$$\mathbb{H}^{n}=\left\{x \in \mathbb{R}^{n : 1} |\langle x, x\rangle_{n : 1}=-1, x_{n+1}>0\right\}$$ <br>
with distance function: <br>
$$
\mathrm{d}_{\mathbb{H}^{n}}(u, v)=\operatorname{arccosh}\left(-\langle u, v\rangle_{n : 1}\right), \quad u, v \in \mathbb{H}^{n}
$$

The tangent space at $p \in \mathbb{H}^{n}$ then inherits the bilinear form of the ambient Minkowski space<br>
<br>
$$
\mathrm{T}_{p} \mathbb{H}^{n}=\left\{x \in \mathbb{R}^{n : 1} |\langle p, x\rangle_{n : 1}=0\right\}
$$

## Optimization in $\mathbb{H}^n$

Compared to the Poincaré disk, the hyperboloid model $\mathbb{H}^n$ has simpler expressions for the gradients and update steps, largely because they can be defined in the ambient space.<br>

For some $p \in \mathbb{H}^n$ we have the exponential map: <br>

$$
\operatorname{Exp}_{p}(v)=\cosh (\|v\|) p+\sinh (\|v\|) \frac{v}{\|v\|} \ \ \ \ \ \ \ \ \text{(3)}
$$ <br>
and $\nabla_{p}^{\mathbb{R}^{n : 1}} E \in \mathbf{T}_{p} \mathbb{R}^{n : 1}$ has the form: <br>
<br>
$$
\nabla_{p}^{\mathbb{R}^{n : 1}} E=\left(\left.\frac{\partial E}{\partial x_{1}}\right|_{p}, \ldots,\left.\frac{\partial E}{\partial x_{n}}\right|_{p},-\left.\frac{\partial E}{\partial x_{n+1}}\right|_{p}\right)
$$ <br>

Gradients of the distance function are straightforward as well: <br>
<br>
$$
\nabla_{u}^{\mathbb{R}^{n : 1}} \mathrm{d}_{\mathbb{H}^{n}}(u, v)=-\left(\langle u, v\rangle_{n : 1}^{2}-1\right)^{-1 / 2} \cdot v \ \ \ \ \ \ \ \ \text{(4)}
$$ <br>

and the final gradient expression: <br>
<br>
$$
\nabla_{p}^{\mathbb{H}^{n}} E=\nabla_{p}^{\mathbb{R}^{n : 1}} E+\left\langle p, \nabla_{p}^{\mathbb{R}^{n : 1}} E\right\rangle_{n : 1} \cdot p \ \ \ \ \ \ \ \ \ \ \ \ \text{(5)}
$$

### Gradient Descent Outline: 
**Input:** Error function $E$ in terms of distance, learning rate $\alpha$, initial value $\Theta = \Theta^{(0)}$ <br>
**Steps:** 
1. Calculate the gradient $\nabla_{\Theta}^{\mathbb{R}^{n+1}} E$ using $(4)$<br>

2. Calculate $\nabla_{\Theta}^{\mathrm{H}^{n}} E$ from $\nabla_{\Theta}^{\mathbb{R}^{n : 1}} E,$ using $(5)$ <br>

3. Set $\Theta^{\text { new }}=\operatorname{Exp}_{\Theta}\left(-\alpha \cdot \nabla_{\Theta}^{\mathrm{H}^{n}} E\right),$ calculated using $(3)$
